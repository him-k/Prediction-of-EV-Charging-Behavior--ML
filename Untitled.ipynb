```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
# loading the csv data to a Pandas DataFrame
ev_data= pd.read_csv('C:\\Users\\Himani\\Desktop\\pp/data.csv')
```

```python
ev_data.isnull().sum()
```

```python
ev_data["donechargingTime"] = ev_data["donechargingTime"].fillna(ev_data["donechargingTime"].mode()[0])
```

```python
NAs = pd.concat([ev_data.isnull().sum()], axis=1, keys=["EV"])
NAs[NAs.sum(axis=1) > 0]
```

```python
ev_data.pop("userID")
```

```python
ev_data.duplicated().sum()
```

```python
for col in ev_data.dtypes[ev_data.dtypes == "object"].index:
    for_dummy = ev_data.pop(col)
    ev_data = pd.concat([ev_data, pd.get_dummies(for_dummy, prefix=col)], axis=1)
ev_data.head()
```

```python
ev_data.drop('Loud Cover',axis=1)
```

```python
X = ev_data.drop(['clusterId','siteId','Loud Cover'],axis=1)
Y = ev_data['Session Length']
```

```python
print(X)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = ev_data['Session Length']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.svm import SVR
```

```python
svm_regressor = SVR(kernel='linear')
```

```python
svm_regressor.fit(X_train, Y_train)
```

```python
Xtrain_pred = svm_regressor.predict(X_train)
```

```python
error_score = metrics.r2_score(Xtrain_pred, Y_train)
print("R squared error of training data : ", error_score)
```

```python
# accuracy score on the test data
X_test_prediction = svm_regressor.predict(X_test)
```

```python
error_score = metrics.r2_score(X_test_prediction, Y_test)
print("R squared error of testing data : ", error_score)
```
